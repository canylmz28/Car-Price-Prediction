# Feature Enginnering

Now, we will drop the unnecessary columns.

```python
columns = ['fuelsystem', 'doornumber', 'aspiration', 'carbody', 'carheight', 'stroke',
        'compressionratio', 'car_ID', 'symboling', 'CarName']
```

```python
data.drop(columns, axis = 1, inplace = True)
```

```python
data.head()
```

```python
data = pd.get_dummies(data, drop_first = True)
```

```python
data.head()
```

# Model Building

```python
import statsmodels.api as sm
```

```python
X = data.drop("price", axis = 1)
y = data["price"]
```

```python
x = sm.add_constant(X)
model = sm.OLS(y, x).fit()
model.summary()
```

enginetype_l has higher p-value, so, we're going to drop it.

```python
x = x.drop("enginetype_l", axis = 1)
model = sm.OLS(y, x).fit()
model.summary()
```

drivewheel_fwd has higher p-value, so, we're going to drop it too.

```python
x = x.drop('drivewheel_fwd', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

boreratio has higher p-value, so, we're going to drop it too.

```python
x = x.drop('boreratio', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

carlength has higher p-value, so we're going to drop it too.

```python
x = x.drop('carlength', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

wheelbase has higher p-value, so, we're going to drop it too.

```python
x = x.drop('wheelbase', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

citympg has higher p-value, so, we're going to drop it too.

```python
x = x.drop('citympg', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

highwaympg has higher p-value, so, we're going to drop it too.

```python
x = x.drop('highwaympg', axis=1)
model = sm.OLS(y, x).fit()
model.summary()
```

 * All the p-values looks just fine. We are good to go.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
model = lr.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

r2_score(y_test, y_pred)
```

```python
sns.regplot(y_test,y_pred)
plt.xlabel("Test Prices", fontsize=14)
plt.ylabel("Prediction Prices", fontsize=14)
plt.title("Test Prices vs. Prediction Prices", fontsize=16);
```
